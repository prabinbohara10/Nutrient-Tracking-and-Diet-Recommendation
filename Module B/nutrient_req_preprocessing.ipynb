```python
def clean_cell(value):
    if pd.isna(value) or not isinstance(value, str):
        return value

    value = re.sub(r'\([^)]*\)', '', value)  # Remove content in parentheses
    value = re.sub(r'[µμ]g', '', value)      # Remove µg/μg
    value = re.sub(r'\sg\s?', ' ', value)    #g with space
    value = re.sub(r'\sg$', '', value)       # g at end
    value = re.sub(r'\smg\s?', ' ', value)   # mg
    value = re.sub(r'\sL\s?', ' ', value)    # L
    value = re.sub(r'\sliters', '', value)   #liters
    
    ## Remove all remaining letters
    value = re.sub(r'[a-zA-Z]', '', value)
    
    ## Clean up spaces
    value = re.sub(r'\s+', ' ', value).strip()
    
    #handling with different dash types
    if "–" in value or "-" in value:
        # Standardize dash
        value = value.replace("–", "-")
        
        # Clean up spaces around dash
        value = re.sub(r'\s*-\s*', '-', value)
        
        # Check if it's a range
        parts = value.split("-")
        if len(parts) == 2:
            # Handle empty parts or parts with only spaces
            parts = [p.strip() if p.strip() else "0" for p in parts]
            return f"{parts[0]}-{parts[1]}"
    
    return value.strip()
```

```python
def clean_df(df):
    
    # Apply cleaning to all data columns
    for col in df.columns[1:]:  #Starting from the 2nd column
        df[col] = df[col].apply(clean_cell)
    
    return df
```

```python
class PatientNutrient:
    def __init__(self, gender, age, bmi):
        self.gender = gender
        self.age = age
        self.bmi = bmi

    def get_nutrient_requirement(self, df):
        age_ranges = [age.split("-") for age in df["Age"]]
        age_group = None
        for i, age_range in enumerate(age_ranges):
            if int(age_range[0]) <= self.age <= int(age_range[1]):
                age_group = df["Age"][i]
        
        if age_group:
            filtered_row = df[(df["Age"] == age_group) & (df['Gender'] == self.gender)]
            return filtered_row.iloc[0].to_dict()
        
        return None
```

```python
class KNN:
    def __init__(self, nutrition_df, nutrition_required: dict):
        self.nutrition_df = nutrition_df
        self.nutrition_required = nutrition_required
        
        self.keys_to_delete = ["Gender", "Age", "Recommended BMI"]

        for key_to_del in self.keys_to_delete:
            if key_to_del in self.nutrition_required:
                del self.nutrition_required[key_to_del]

        self.index_distance_sorted = {}
        self.__create_index_distance_map()
    
    def __create_index_distance_map(self):
        index_distance = {}
        for index, row in self.nutrition_df.iterrows():
            row_dict = row.to_dict()

            distance = self.__calculate_distance(ref_dict= row_dict, required_dict= self.nutrition_required)
            index_distance[str(index)] = distance

        # print(index_distance)
        # for k, v in index_distance.items():
        #     print(k," - ", v)
        self.index_distance_sorted = sorted(index_distance.items(), key= lambda x: x[1])

    def __calculate_distance(self, ref_dict, required_dict):
        distance = 0
        for key, value in required_dict.items():
            value = value.strip("-")[0]
            sq = (float(value) - float(ref_dict[key])) ** 2
            distance += sq
        
        distance = pow(distance, 0.5)

        return distance
    
    def get_food(self, k = 4):
        '''
        Return top k food list
        '''
        top_k_index = [idx for idx, dist in self.index_distance_sorted[:k]]
        
        return self.nutrition_df["food"].iloc[top_k_index]

```
